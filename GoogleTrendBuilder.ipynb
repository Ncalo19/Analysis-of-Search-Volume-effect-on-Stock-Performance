```python
import time
import json
import numpy as np
import pandas as pd
from pytrends.request import TrendReq
```

```python
with open('sp500_companies.json') as f:
    companies = json.loads(f.read())
symbol = [i['Symbol'] + ' stock' for i in companies]
print(len(symbol))

sp500_trends = []
```

```python
pytrends = TrendReq(hl='en-US', tz=360)
```

```python
def recursive_trend_call(index=0):
    for i in np.arange(index, len(symbol)):
        if ((i%10) == 0):
            print('{} / 505 Complete'.format(i))
        try:
            pytrends.build_payload([symbol[i]], cat=0, timeframe='today 5-y', geo='', gprop='')
            sp500_trends.append(pytrends.interest_over_time())
            time.sleep(1)
        except Exception as e:
            print(e)
            print('Exception hit... going to sleep.')
            time.sleep(100)
            print('Awake and recursing')
            recursive_trend_call(index=i)
    print('COMPLETE!')
```

```python
recursive_trend_call()
```
