```python
merged_data = pd.merge(relevant_earnings, stocks, how='right', left_on=['date', 'symbol'], right_on =['date', 'symbol'])
pd.set_option('display.max_rows', 500)
merged_data
```

```python
merged_data['earnings_difference'] = merged_data['eps'] - merged_data['eps_est']
merged_data
```

# Regression and correlation

```python
merged_data.corr()
```

```python
import matplotlib.pyplot as plt
x = merged_data['sum_search_value']
y = merged_data['earnings_difference']
plt.plot(x, y, 'o')

from numpy.polynomial.polynomial import polyfit
m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x + b)
plt.show()
```

```python
import seaborn as sns
image = sns.regplot(x,y,ci=None,data=merged_data, scatter_kws={"color": "black"}, line_kws={"color": "red"})
image.axhline(0, label="Y axis")
plt.xlabel("Search Popularity")
plt.ylabel("Difference between Estimate and Earnings")
plt.legend()
plt.title("Ear")
```

```python
from sklearn.linear_model import LinearRegression
x = np.nan_to_num(x)
y = np.nan_to_num(y)
reg = LinearRegression().fit(x.reshape(-1, 1),y.reshape(-1, 1))
reg.score(x.reshape(-1, 1), y.reshape(-1, 1))
```
